```python
# Vẽ hình vuông bằng vòng lặp for
import turtle

p = turtle.Turtle()
p.pensize(5)
p.pencolor("green")
a = int(input("Nhập kích thước cạnh của hình vuông : "))
for i in range(4):
    p.fd(a)
    p.lt(90)
turtle.done()
```

```python
# Sử dụng vòng lặp lồng nhau để vẽ hình 
import turtle

p = turtle.Turtle()
p.pensize(1)
p.pencolor("green")
p.speed(20)

for i  in range(1,73):
    p.lt(5)
    for i  in range(1,6):
        p.fd(200)
        p.lt(144)
turtle.done() 
```

```python
# Vẽ hình chữ nhật bằng ký tự
char = input("Nhập ký tự bạn muốn vẽ: ")
height = int(input("Nhập chiều cao của hình chữ nhật: "))
width = int(input("Nhập vào chiều rộng hình chữ nhật: "))

for i in range(1, height + 1):
    print_str = ""
    for j in range(1, width + 1):
        if i == 1 or i == height:   
            print_str += char
        else:
            if j == 1 or j == width:
                print_str += char
            else:
                print_str += ' '
    print(print_str)

        
```

```python
# Fizz Buzz
a = int(input("Nhập a: "))
b = int(input("Nhập b: "))
if a > b:
    print("Số kết thúc cần lớn hơn số bắt đầu")
else:
    for i in range(a, b + 1):
        if not i % 15:
            print("Fizz Buzz")
        elif not i % 3:
            print("Fizz")
        elif not i % 5:
            print("Buzz")
        
        else:
            print(i)
```

```python
# The 21 Game 
import random as r 



while True:  
    current_number = 1 
    if r.randint(0,1) == 0: 
        player = "human"
    else:
        player = "computer"
    
    
    while current_number <= 21:
        print(f"Current total is {current_number}")
        if player == "human":
            player_choice = 0
            while player_choice not in [1, 2, 3]:
                player_choice = int(input("Your turn, Add 1 or 2 or 3 into total, Don't pass 21, you will lose: "))
            
            current_number = current_number + player_choice
            
            if current_number >= 21:
                print(f"The current number is {current_number}")
                print("You lose")
                break
            player = "computer"
        else:
            computer_choice = r.randint(1,3)
            print(f"Computer turn, Computer choise is {computer_choice}")
            current_number += computer_choice
            if current_number >= 21:
                print(f"The current number is {current_number}")
                print("You won")
                break
            player = "human"
    break
print("goodbye")
```
