```python
#Vẽ hình tròn
import turtle
pen = turtle.Turtle()
pen.pensize(10)

pen.pencolor("blue")

pen.fillcolor("green")
pen.begin_fill()

pen.circle(100)
pen.end_fill()
turtle.done()
```

```python
# Vẽ hình chữ nhật
import turtle
pen = turtle.Turtle()
pen.pensize(5)
pen.pencolor("red")
pen.fillcolor("gray")
pen.begin_fill()
pen.forward(200)
pen.left(90)
pen.forward(200)
pen.left(90)
pen.forward(200)
pen.left(90)
pen.forward(200)
pen.end_fill()
turtle.done()
```

```python
# Vẽ ngôi nhà mơ ước
import turtle
pen = turtle.Turtle()
turtle.bgcolor("#01DFD7")
pen.pensize(2)
pen.fillcolor("yellow")
pen.begin_fill()
pen.rt(90)
pen.fd(200)
pen.rt(70)
pen.fd(150)
pen.rt(110)
pen.fd(200)
pen.rt(70)
pen.fd(150)
pen.bk(150)
pen.lt(160)
pen.fd(200)
pen.lt(90)
pen.fd(200)
pen.lt(90)
pen.fd(200)
pen.lt(90)
pen.fd(200)
pen.end_fill()

pen.fillcolor("red")

pen.begin_fill()
pen.lt(60)
pen.fd(200/3**0.5)
pen.lt(60)
pen.fd(200/3**0.5)
pen.lt(150)
pen.fd(200)
pen.lt(150)
pen.fd(200/3**0.5)
pen.rt(130)
pen.fd(150)
pen.rt(50)
pen.fd(200/3**0.5)
pen.end_fill()

pen.fillcolor("orange")
pen.begin_fill()
pen.penup()
pen.goto(200,200)
pen.pendown()
pen.circle(50)
pen.end_fill()

pen.fillcolor("green")
pen.begin_fill()
pen.penup()
pen.goto(100,-200)
pen.lt(30)
pen.pendown()
pen.fd(200)
pen.lt(150)
pen.fd(200/3**0.5)
pen.rt(150)
pen.fd(100)
pen.lt(150)
pen.fd(200/3**0.5)
pen.rt(150)
pen.fd(100)
pen.lt(150)
pen.fd(200/3**0.5)
pen.lt(60)
pen.fd(200/3**0.5)
pen.lt(150)
pen.fd(100)
pen.rt(150)
pen.fd(200/3**0.5)
pen.lt(150)
pen.fd(100)
pen.rt(150)
pen.fd(200/3**0.5)
pen.lt(150)
pen.end_fill()

pen.fillcolor("brown")
pen.begin_fill()
pen.fd(120)
pen.rt(90)
pen.fd(60)
pen.rt(90)
pen.fd(40)
pen.rt(90)
pen.fd(60)
pen.lt(110)
pen.end_fill()

pen.fillcolor("blue")
pen.penup()
pen.goto(0,-200)
pen.pendown()
pen.fd(150)
pen.rt(20)
pen.fd(70)
pen.rt(90)

pen.begin_fill()
pen.fd(100)
pen.lt(90)
pen.fd(60)
pen.lt(90)
pen.fd(100)
pen.end_fill()

pen.penup()
pen.goto(-30,-100)
pen.pendown()
pen.begin_fill()
pen.rt(70)
pen.fd(40)
pen.rt(110)
pen.fd(50)
pen.rt(70)
pen.fd(40)
pen.rt(110)
pen.fd(50)
pen.rt(70)
pen.fd(40)
pen.end_fill()

turtle.done()
```
