```python
# Vẽ hình vuông sử dụng hàm
import turtle
p = turtle.Turtle()
p.pencolor("green")

def square(a):
    for i in range(4):
        p.fd(a)
        p.lt(90)
    turtle.done()
square(200)
```

```python
# Vẽ đa giác đều
import turtle
p = turtle.Turtle()
p.pencolor("green")

def polygons(n, width):
    angle = (n -2) * 180 / n
    for i in range(n):
        p.fd(width)
        p.lt(180 - angle)
    turtle.done()
polygons(9, 100)
```

```python
#Chuyển đổi nhiệt độ (F -> C)
def trans(f):
    f -= 32
    f /= 1.8
    return f
print(f"{trans(182)} độ  C") 
```

```python
# Tính diện tích hình tròn và vẽ
import turtle
import math

def draw_circle(r):
    p = turtle.Turtle()
    p.pencolor("green")
    p.circle(r)
    turtle.done()
def calc_circle(r):
    return round(math.pi * r * r, 2)
r = float(input("Nhập bán kính hình tròn: "))
draw_circle(r)
print(f"Diện tích hình tròn là: {calc_circle(r)}")
```

```python
# Vẽ hàm sử dụng hàm

import  turtle
def draw_house():
    p = turtle.Turtle()
    p.pencolor("green")
    p.fillcolor("orange")
    p.begin_fill()
    p.lt(90)
    p.fd(150)
    p.lt(90)
    p.fd(100)
    p.lt(90)
    p.fd(150)
    p.rt(90)
    p.bk(100)
    p.fd(200)
    p.rt(90)
    p.fd(300)
    p.rt(90)
    p.fd(200)
    p.rt(90)
    p.fd(150)
    p.rt(90)
    p.fd(200)
    p.bk(100)
    p.rt(90)
    p.fd(150)
    p.bk(300)
    p.rt(90)
    p.end_fill()
    
    p.fd(100)
    p.lt(90)
    p.fillcolor("gray")
    p.begin_fill()
    
    p.fd(150)
    p.rt(90)
    p.fd(150)
    p.rt(90)
    p.fd(150)
    p.rt(90)
    p.fd(40)
    p.rt(90)
    p.fd(100)
    p.lt(90)
    p.fd(70)
    p.lt(90)
    p.fd(100)
    p.lt(90)
    p.fd(70)
    p.bk(35)
    p.lt(90)
    p.fd(100)
    p.bk(100)
    p.lt(90)
    p.fd(75)
    p.end_fill()
    
    p.penup()
    p.goto(-30, 50)
    p.pendown()
    p.fillcolor("white")
    p.begin_fill()
    p.fd(40)
    p.rt(90)
    p.fd(50)
    p.rt(90)
    p.fd(40)
    p.rt(90)
    p.fd(50)
    p.rt(90)
    
    p.penup()
    p.goto(-130, 50)
    p.pendown()
    p.fd(40)
    p.rt(90)
    p.fd(50)
    p.rt(90)
    p.fd(40)
    p.rt(90)
    p.fd(50)
    p.rt(90)
    p.end_fill()
    
    p.begin_fill()
    p.penup()
    p.goto(-130, 200)
    p.pendown()
    p.fd(40)
    p.rt(90)
    p.fd(50)
    p.rt(90)
    p.fd(40)
    p.rt(90)
    p.fd(50)
    p.rt(90)
    p.end_fill()
    
    p.begin_fill()
    p.penup()
    p.goto(-30, 200)
    p.pendown()
    p.fd(40)
    p.rt(90)
    p.fd(50)
    p.rt(90)
    p.fd(40)
    p.rt(90)
    p.fd(50)
    p.rt(90)
    p.end_fill()
    
    turtle.done()
draw_house()
```

```python
# Vẽ ngôi nhà mơ ước sử dụng hàm
import turtle
def draw_dream_house():
    
    pen = turtle.Turtle()
    pen.pensize(5)
    a = 25 * 2 ** 0.5
    pen.pencolor("blue")
    pen.fd(250)
    pen.lt(90)
    pen.fd(150)
    pen.lt(90)
    pen.fd(250)
    pen.lt(90)
    pen.fd(150)
    pen.lt(90)

    pen.fd(90)
    pen.lt(90)
    pen.fd(100)
    pen.rt(90)
    pen.fd(70)
    pen.rt(90)
    pen.fd(100)
    pen.lt(90)
    pen.fd(90)
    pen.lt(90)
    pen.fd(150)
    pen.lt(90)
    pen.fd(250)

    pen.rt(135)
    pen.fd(150)
    pen.rt(90)
    pen.fd(50)
    pen.lt(135)
    pen.fd(a)
    pen.penup()
    pen.fd(20)
    pen.pendown()
    pen.fd(a)

    pen.penup()
    pen.rt(90)
    pen.fd(a)
    pen.pendown()
    pen.rt(90)
    pen.fd(a)
    pen.penup()
    pen.fd(20)
    pen.pendown()
    pen.rt(90)
    pen.fd(a)
    pen.bk(a)
    pen.lt(90)
    pen.fd(2*a)
    pen.rt(135)
    pen.fd(50)
    pen.bk(100)
    
    pen.pencolor("black")
    pen.fillcolor("green")
    pen.begin_fill()
    pen.penup()
    pen.goto(-300,-50)
    pen.rt(135)
    pen.pendown()
    pen.fd(200)
    pen.lt(150)
    pen.fd(200/3**0.5)
    pen.rt(150)
    pen.fd(100)
    pen.lt(150)
    pen.fd(200/3**0.5)
    pen.rt(150)
    pen.fd(100)
    pen.lt(150)
    pen.fd(200/3**0.5)
    pen.lt(60)
    pen.fd(200/3**0.5)
    pen.lt(150)
    pen.fd(100)
    pen.rt(150)
    pen.fd(200/3**0.5)
    pen.lt(150)
    pen.fd(100)
    pen.rt(150)
    pen.fd(200/3**0.5)
    pen.lt(150)
    pen.end_fill()

    pen.fillcolor("brown")
    pen.begin_fill()
    pen.fd(120)
    pen.rt(90)
    pen.fd(60)
    pen.rt(90)
    pen.fd(40)
    pen.rt(90)
    pen.fd(60)
    pen.lt(110)
    pen.end_fill()
    turtle.done()
draw_dream_house()
```
