```python
from tkinter import *
window = Tk()
window.geometry("400x300")
window.eval('tk::PlaceWindow . center')

window.title("Simple calculater")
num1 = Entry(window)
num2 = Entry(window)

result_lbl = Label(window, text="Result")

def add():
    result = int(num1.get()) + int(num2.get())
    result_lbl.config(text=result)
def sub():
    result = int(num1.get()) - int(num2.get())
    result_lbl.config(text=result)
def mul():
    result = int(num1.get()) * int(num2.get())
    result_lbl.config(text=result)
def div():
    result = round(int(num1.get()) / int(num2.get()), 2)
    result_lbl.config(text=result)


add_btn = Button(window, text="+", width=5, command=add)
sub_btn = Button(window, text="-", width=5, command=sub)
mul_btn = Button(window, text="*", width=5, command=mul)
div_btn = Button(window, text="/", width=5, command=div)



num1.pack() 
num2.pack() 
add_btn.pack() 
sub_btn.pack() 
mul_btn.pack() 
div_btn.pack() 
result_lbl.pack() 

window.mainloop()
```

```python
from tkinter import *
window = Tk()
window.geometry("400x300")
window.eval('tk::PlaceWindow . center')
window.title("My Intro")

output_lbl = Label(window, text="Output")
firstname_lbl = Label(window, text="Fisrtname")
lastname_lbl = Label(window, text="Lastname")

firstname = Entry(window)
lastname = Entry(window)

def subm():
    out = "My name is " + lastname.get() +" " + firstname.get()
    output_lbl.config(text=out)
    
    

submit_btn = Button(window, text="Submit", command=subm)


    
output_lbl.grid(column=0, row=0, columnspan=2)
firstname_lbl.grid(column=0, row=1)
lastname_lbl.grid(column=0, row=2)
firstname.grid(column=1, row=1)
lastname.grid(column=1, row=2)
submit_btn.grid(column=0, row=3, columnspan=2)


window.mainloop()
```

```python
from tkinter import *
from tkinter import messagebox
window = Tk()
window.geometry("400x300")
window.eval('tk::PlaceWindow . center')
window.title("My BMI")

output_lbl = Label(window, text="BIM calculator")
weight_lbl = Label(window, text="My weight")
height_lbl = Label(window, text="My height")

weight = Entry(window)
height = Entry(window)

def calc():
    bim = float(weight.get()) / float(height.get())
    messagebox.showinfo("BIM result", f"Your BIM is {bim}")   
    

submit_btn = Button(window, text="Submit", command=calc)


    
output_lbl.grid(column=0, row=0, columnspan=2)
weight_lbl.grid(column=0, row=1)
height_lbl.grid(column=0, row=2)
weight.grid(column=1, row=1)
height.grid(column=1, row=2)
submit_btn.grid(column=0, row=3, columnspan=2)


window.mainloop()
```
